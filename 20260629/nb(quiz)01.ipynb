```python
# 문제10) 가입연수별 그룹 분석
# 가입연수를 기준으로 다음과 같이 그룹을 나누시오.

# 0~2년
# 3~5년
# 6년 이상

# 각 그룹의 평균 구매금액을 구하시오.


# 8. 0~2년, 3~5년, 6년 이상 각 그룹의 평균 구매금액을 구하시오.

price1 , price2, price3= 0, 0, 0 # 그룹별 구매금액 누적 price1, price2, price3 = (0. 0. 0)
count1, count2, count3 = 0, 0, 0 # 그룹별 카운트 누적


for o in orders:
    y = o["가입연수"]

    if y <= 2: # 2년이하
        price1 = price1 + o["구매금액"]
        count1 = count1 + 1

    elif y <= 5:
        price2 = price2 + o["구매금액"]
        count2 = count2 + 1

    else: # 6년이상
        price3 = price3 + o["구매금액"]
        count3 = count3 + 1

print(f"0~2년: {price1 / count1:,}")
print(f"3~5년: {price2 / count2:,}")
print(f"6~년: {price3 / count3:,}")
```

```python
#[심화 문제]: 딕셔너리, 리스트, 반복문, 조건문을 모두 활용
#문제11) 가장 구매금액이 높은 고객 TOP 3를 출력하시오.
#11-2) 연평균 구매금액(소수점1자리ㅠ반올림)이 높은 구매고객 TOP3?

#9. Top3 (예제)

result = []

for o in orders:
    result.append(o)

# bubble sort (단순 버전)
for i in range(len(result)):
    for j in range(len(result)):
        if result[i]["구매금액"] > result[j]["구매금액"]:
            temp = result[i]
            result[i] = result[j]
            result[j] = temp

print(result[0])
print(result[1])
print(result[2])
```
