```python
data= []

file=open('9-4)_work.csv', 'r', encoding='utf-8')
lines = file.readlines()

for line in lines[1:]:
    name,kor,eng=line.strip().split(",")

    data.append({
          "이름": name,
          "국어": int(kor),
          "영어": int(eng),
          })
    
file.close()

names = [student['이름'] for student in data]
kor_scores = [student['국어'] for student in data]
eng_scores = [student['영어'] for student in data]

name_kor = list(zip(names, kor_scores))
name_eng = list(zip(names, kor_scores))

print(name_kor)
print(name_eng)

name_kor_dict = dict(zip(names, kor_scores))
name_eng_dict = dict(zip(names, eng_scores))

print(name_kor_dict)
print(name_eng_dict)

avg_list = list(map(
    lambda x: (
        x['이름'], 
        (x['국어'] + x['영어']) / 2
         ),
         data
         ))

sorted_avg = sorted(
    avg_list,
    key=lambda x: x[1],
    reverse=True
)

top_students = list(filter(
    lambda x: x[1] >=90,
    sorted_avg
))

print(sorted_avg)
print(top_students)
```
