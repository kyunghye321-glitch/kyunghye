```python
result = {}

for r in sales:
    p_name = r['상품']
    price = int(r['판매액'])
    total_cnt = r['수량']
    if p_name not in result:
        result[p_name] = {'합계':0, '건수':0, '최대':price, '최소':price, '총수량':0}
        # {
        #     '키보드':{'합계':0, '최대':0, '최소':0, '총수량':0}
        #     '모니터':{'합계':0, '최대':0, '최소':0, '총수량':0}
        # }
    info = result[p_name]
    info['합계'] += price
    info['건수'] += 1
    info['총수량'] += total_cnt

    # 최대/최소
    # result['노트북']['최대']
    if info['최대'] < price:
        info['최대'] = price

    if info['최소'] > price:
        info['최소'] = price

print(result)
```
