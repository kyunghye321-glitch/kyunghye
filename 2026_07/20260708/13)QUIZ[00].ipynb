```python
# 실습 과제 1. 누적합(Prefix Sum)

steps = [5231,6120,4890,7005,6500,7200,6800]

total = 0

for i in range(len(steps)):
    total = total + steps[i]
    print(i + 1, '일 누적:', total)
```

```python
# 실습 과제 2. 누적합(Prefix Sum)

steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]

total = 0

for day in steps:
    for date in day:
        total = total + day[date]
        print(date, '누적: ', total)
```

```python
# 실습 과제 3. 슬라이딩 윈도우

sales = [12,15,18,20,17,25,22,19]

window = sales[0] + sales[1] + sales[2]
max_val = window

print(f"{sales[0]} + {sales[1]} + {sales[2]} = {window}")

for i in range(1, len(sales) - 2):
    window = window - sales[i - 1] + sales[i + 2]
    print(f"{sales[i]} + {sales[i + 1]} + {sales[i + 2]} = {window}")

if window > max_val:
    max_val = window


print()
print('최대 매출 합: ', max_val)
```
