```python
def hello():
    """
    '안녕'을 출력하고 None을 반환하는 함수
    """
    print('안녕')

result = hello()
# 함수호출은 표현식이다. 표현식은 값을 남긴다. 반환값이 없는 함수를 호출하면 None값을 남긴다.
print(result)
# result에는 None값이 할당되어 있다.
```

```python
def add(a, b):
    return a+b

print(add(3, 5))
result = add(3,5)
print(result)
```

```python
def calc(a, b):
    x = a + b
    y = a - b
    z = a * b
    return x, y, z #3개의 값을 반환하는 것이 아니라 tuple로 묶어서 반환

result2 = calc(10, 5) 
print(result2)
```

```python
r1, r2, r3 = calc(10, 5) # 언패킹
print(r1)
print(r2)
print(r3)
```

```python
def greet(name):
    print(f"{name}님 안녕하세요")

greet('철수')
greet('영희')

#함수 안에 매개변수도 함수 코드블록 안에서만 사용가능한 지역변수이다
print(name) # NameError: name 'name' is not defined
```

```python
def fun1(a, b):
    print('매개변수 a의 인수: ', a)
    print('매개변수 b의 인수: ', b)

# 함수 호출 시 인자값(인수)의 위치로 매개변수 값 결정
fun1(1, 2)
fun1(10, 20)

# 함수 호출 시 매개변수 이름을 지정하여 인자값(인수) 입력
fun1(a=100, b=200)
fun1(a=777, b=999)
# 매개변수 이름을 지정하여 인수 입력 시(키워드 인수) 순서는 상관없음
```

```python
def greet(name, age=20):
    print(name, age)

greet('홍길동')
greet('이순신', 30)
```

```python
# 오류 발생: SyntaxError: parameter without a default follows parameter with a default
def greet(name='홍길동', age=20):
    print(name, age)
```
