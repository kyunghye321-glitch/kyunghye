```python
# zip()
# zip()은 여러 개의 시퀀스 객체를 같은 위치(인덱스)의 값끼리 묶어주는 함수이다.
# zip -> list

names = ['홍길동', '이순신', '김유신']
scores = [90, 80, 100, 80]
gender = ['F', 'M', 'F']
result = zip(names, scores, gender)
print(list(result))
```

```python
# zip -> dict

names = ['홍길동', '이순신', '김유신']
scores = [90, 80, 100, 80]
gender = ['F', 'M', 'F']

result = dict(zip(names, scores, gender))
print(result)
```
