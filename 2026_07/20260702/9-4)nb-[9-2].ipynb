```python
# 1급 함수(First-Class Function)​란 함수를 일반 데이터(객체)처럼 다룰 수 있는 함수를 말한다.
# 파이썬에서는 함수도 객체이므로 1급 함수를 지원한다.
# 즉, 함수를 다음과 같이 사용할 수 있다.


# 1) 변수에 저장할 수 있다. 
# 2) 함수의 인수(argument)로 전달할 수 있다. 
# 3) 함수의 반환값(return)으로 사용할 수 있다. 
# 4) 자료구조(list, dict 등)에 저장할 수 있다.
```

```python
# 변수에 저장하기

def hello():
    print("Hello")

f = hello      # 함수 자체를 저장 (괄호 없음)

f()
```

```python
# 함수의 인수로 전달하기

def square(x):
    return x * x

def test(fun):
    return fun(5)

print(test(square))
```

```python
# 함수를 반환하기

def outer():
    def inner():
        print("inner")

    return inner

f = outer()

f()

(outer())()
```

```python
# 자료 구조에 저장하기

def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))
```

```python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

calc = {
    '+': add,
    '-': sub
}

print(calc['+'](10, 5))
print(calc['-'](10, 5))
```

```python
def plus(x):
    return x + 10

a = lambda x: x + 10
# 변수 전장 -> 1급 함수 기능 포함
```
