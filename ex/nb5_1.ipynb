```python
## f-string
## 객체를 문자열포맷으로 출력하는 포맷팅 문법. 모든 객체를 문자열로 변경 후 포맷팅

```

```python
# 1) 문자 채우기, 정렬
name = "Python"

print(f"{name:*<15}")  # 왼쪽 정렬, 빈칸은 *로 채움
print(f"{name:*^15}")  # 가운데 정렬, 빈칸은 *로 채움
print(f"{name:*>15}")  # 오른쪽 정렬, 빈칸은 *로 채움
# 변수 사용 가능
w = 15
print(f"{name:*>{w}}")  # 오른쪽 정렬, 빈칸은 *로 채움
```

```python
# 2) 숫자 채우기
num = 123

print(f"{num:08}")   # 전체 8자리 문자로 변경 후, 부족한 자리는 0으로 채움
print(f"{num:010}")
```

```python
# 3) 부호 포함 출력
a = 150
b = -50

print(f"{a:+}")  # 문자열 변경시 양수는 +, 음수는 - 표시
print(f"{b:+}")    # 음수는 - 표시
```

```python
# 4) 양수는 공백, 음수만 부호 포함 출력
c = 100
d = -100

print(f"{c: }")   # 양수 앞에 공백
print(f"{d: }")   # 음수는 - 유지
```

```python
# 5) 천 단위 구분기호 출력
sales = 1234567890
print(f"{sales:,}")  # 천 단위마다 콤마 추가
```

```python
# 6) 천 단휘 구분기호 + 자릿수 맞춤
sales2 = 1234567
print(f"{sales2:>15,}")  # 오른쪽 정렬 + 천 단위 구분
print(f"{sales2:<15,}")  # 오른쪽 정렬 + 천 단위 구분
```

```python
# 7) 소수점 반올림
pi = 3.1415926535
print(f"{pi:.2f}")   # 소수점 둘째 자리까지 반올림 후 문자열 출력
print(f"{pi:.4f}")   # 소수점 넷째 자리까지
```

```python
# 8) 지수 표기법
num = 123456789
print(f"{num:e}")  # 소문자 e 
# 1.234568e+08
print(f"{num:E}")  # 대문자 E
# 1.234568E+08
```

```python
# 9) 백분율 표시
rate = 0.8746
print(f"{rate:.0%}")  # 100을 곱하고 반올림
print(f"{rate:.1%}")  # 100을 곱하고 소수점 첫째 자리까지 반올림 % 표시
print(f"{rate:.2%}")

# 87%
# 87.5%
# 87.46%
```

```python
# 10) 2진수, 8진수, 16진수 
num = 255
print(f"{num:b}")  # 11111111 2진수
print(f"{num:o}")  # 377      8진수
print(f"{num:x}")  # ff       16진수 소문자
print(f"{num:X}")  # FF       16진수 대문자
```

```python
# 11) 날짜 포맷팅
# 파이썬 언어에 포함된 객체/함수라도 내장객체/함수와 모듈안에 포함된 객체/함수로 구분된다.
# 모듈안에 포함된 것들은 import가 필요하다. 뒤에서 모듈수업때 설명
from datetime import datetime # datetime 모듈(module) 안에 있는 datetime 클래스(객체)를 가져온다.
today = datetime.now()
print(f"{today:%Y-%m-%d}")          # 연-월-일
print(f"{today:%H:%M:%S}")          # 시:분:초
print(f"{today:%Y/%m/%d %H:%M}")    # 날짜 + 시간
# 2026-06-22
# 19:48:14
# 2026/06/22 19:48
```

```python
# 12) 변수와 값을 같이 출력
name = "Kim"
age = 20
print(f"{name=}")  # 변수명=값
print(f"{age=}")
```

```python
# 13) 계산결과
a = 10
b = 20
print(f"{a} + {b} = {a+b}")       # 계산식 결과 출력
print(f"{a*b=}")                  # 변수명 포함 출력
```

```python
# 14) 표형태로 출력
name = "홍길동"
age = 25
score = 95.5

print(f"{'이름':<10}{'나이':<10}{'점수':<10}")
print(f"{name:<10}{age:<10}{score:<10.1f}")
```

```python
# 15) 복합
money = 3500000.75
print(f"{money:>15,.2f}")
# > : 오른쪽 정렬
# 15 : 전체 폭 15칸
# , : 천 단위 콤마
# .2f : 소수점 2자리 실수
```

```python
# 16) 보고서 형태
income = 123456789
expense = 98765432
profit = income - expense

print(f"{'매출':<10}: {income:>15,}")
print(f"{'비용':<10}: {expense:>15,}")
print(f"{'이익':<10}: {profit:>15,}")
```

```python
##변수
# 1) 객체가 처음으로 할당될 때 생성. 추가 할당은 이미 생성된 변수의 참조를 변경.
# 2) 객체를 가리키는 링크를 파이썬에선 참조(reference)라 부른다.
# 3) 어떠한 타입의 객체도 참조할 수 있으며, 변수는 사용되기 전에 할당되어야 한다.
# 4) 객체는 자신이 의미하는 값과 기능을 가진 할당된 메모리 조각이다.

a = 3
a = "hello"

# 1) 숫자 3을 나타내는 새로운 객체 생성
# 2) 변수 a가 아직 존재하지 않으면 생성
# 3) 변수 a를 새로운 객체 3으로 연결(대입, 할당, 참조)
# 4) 문자열 "hello"를 나타내는 새로운 객체 생성
# 5) 변수 a를 객체 "hello"로 연결

# 공유 참조
# 불변 객체와 가변 객체는 공유 참조(Shared Reference)가 발생했을 때 동작이 다르다.
# 1) 불변 객체의 공유 참조: int, float, bool, str, tuple, frozenset 등이 불변 객체

a = 10
b = a

print(a, b) # 10 10
print(a is b)  # True 같은 객체 참조
print(a == b)

#같은 객체를 참조: 공유참조
#숫자는 불변 객체이므로 값 변경이 아니라 새로운 20을 만들고 a는 20을 참조
#:불변객체의 공유참조 값 변경시 새로운 객체를 생성 후 참조


#문자열도 불변 객체
s1 = "Python"
s2 = "Pytho"

s2 = s2+"n"

print(s1 is s2) # True
print(s1 == s2) # True
```

```python
#가변 객체의 공유 참조
#list, dict, set, 사용자 정의 클래스가 가변 객체

nums1 = [10, 20, 30]
nums2 = nums1

print(nums1 is nums2) # True

nums1[0] = 999
print(nums1, nums2)

nums1.append(40)

print(nums1) # [10, 20, 30, 40]
print(nums2) # [10, 20, 30, 40]
```

```python
# ​등가(같은 값) 비교(==)와 동일 객체(같은 참조) 비교(is)
# 불변 객체의 경우 같은 값의 객체로 변수를 대입하면 파이썬 내부 메커니즘으로(값수정이 안되는 불변객체를 메모리 안에 여럿 만드는 것은 메모리 낭비) 동일한 객체를 참조하게 되어있다.

a = 100
b = 100

print(a == b)  # True 값 비교
print(a is b)  # True 객체 비교

# 내장 함수 id()를 이용하면 동일 객체를 참조하는지 확인 가능
print(id(a))
print(id(b))
```
