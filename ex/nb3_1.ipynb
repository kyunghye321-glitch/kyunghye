```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))
print(3>5)
```

```python
# 5. 문자열 (str)
e = "Python"
print(e)
print(type(e))

e = "PKthon"
print(e[1])

#e[1]: 'K' error 문자열 객체가 불변
```

```python
# 6. 리스트 (list)
f = [1, 2, 3]
print(f)
print(type(f))

print(f[1])
f[1]='abc'
print(f)
```

```python
# 7. 튜플 (tuple): 불변
g = (1, 2, 3)
print(g)
print(type(g))

print(g[1])
```

```python
# 8. 딕셔너리 (dict): 가변
h = {"id": "admin", "pw": "1234"}
print(h)
print(type(h))

print(h["id"])

h["pw"] = "0000"
print(h)
```

```python
# 9. 집합 (set)| 중복X, 개별X
i = {1, 2, 3}
print(i)
print(type(i))
```

```python
# 10. 범위 (range)
j = range(5)
print(j)
print(type(j))
print(list(j))
```

```python
# 14. 함수 (function)
def hello():
    print("Hello") 

print(hello)
print(type(hello))
```
