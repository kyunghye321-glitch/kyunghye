```python
# 파이썬의 내장 객체중 하나의 값만 가지는 객체(숫자, bool)는 그 값을 변경X (immutable : 불변)
# a = 10
# a = 20

# a객체가 가진 10라는 값을 20으로 변경하는 것이 아니라 20이라는 새로운 객체를 메모리에 만들고 a가 20을 참조하게 하는 것
# 파이썬의 문자열도 변경X (immutable : 불변)

s = "python"
print(s[0])
s[0] = "C" # error

# 파이썬 타입 중 list, dict, set은 가변 타입(mutable) [객체 생성 후 내부 데이터를 변경할 수 있는 객체를 말한다]
nums = [1, 2, 3]
nums[0] = 100
print(nums)

#문자열은 변경할 수 없기 때문에 문자열의 연산은 새로운 문자열을 생성 후 반환
```

```python
print('python')
print("python")

# 여려줄 문자열

text = '''
안녕하세요
파이썬입니다
'''
print(text)

text = """
안녕하세요
파이썬입니다
"""
print(text)

# 이스케이프 문자 
print("Hello\nPython")

# Raw 문자열 : 문자열 안의 이스케이프 문자(\n, \t, \\ 등)를 특별하게 해석하지 않고 그대로 표현
print(r"C:\Users\data")

# f-string : 문자열 안에 변수 표현식{변수} 사용
name = "Tom"
print(f"이름은 {name}이다")
print("이름은 "+name+"이다")

age = 20
print("나이는 "+str(age)+"입니다")
print(f"나이는 {age}이다")
#문자열 포맷 방식
```

```python
# 함수 & 메서드
# 함수: 전역에서 정의된 독립적인 객체. 대부분 다른 객체를 매개값(인자)으로 받아 어떤 기능을 함.
# 메서드: 특정객체에 소속된 함수. 멤버함수 or 메서드라 부르며 해당객체에 대해서만 어떤 기능을 하는 함수

def fx_function():
    print('hello_python!')

fx_function()
fx_function()
fx_function()
fx_function()
# >> DRY(Don't Repeat Yourself)
```

```python
# dir 함수: dir함수는 인자로 전달된 객체에서 이용 가능한 모든 속성을 반환

s = "python"
print(dir(s))
print(s.upper())
print(s)
#대부분의 객체 메서드는 자기자신은 수정: 문자열은 불변이기 때문
```

```python
#대소문자 변환
text = "Python Programming"
print(text.upper())      # 모두 대문자
print(text.lower())      # 모두 소문자
print(text.capitalize()) # 첫 글자만 대문자
print(text.title())      # 각 단어 첫 글자 대문자

#공백제거
text = "   Python   "
print(text.strip())   # 양쪽 공백 제거
print(text.lstrip())  # 왼쪽 공백 제거
print(text.rstrip())  # 오른쪽 공백 제거
```

```python
#문자열 검색
text = "Python Programming"

print(text.find("Pro"))
print(text.find("Java")) # -1을 반환하면 글자를 찾지 못했을 때
print(text.startswith("Python")) # 성이 "김" name, startswith("김")
print(text.endswith("ing")) #파일 중에 한글 파일만 f.emdswith(".hwp")

#문자열 치환
text = "I like Java"
print(text.replace("Java", "Python"))
```

```python
#문자열 분리
data = "apple,banana,orange"
result = data.split(",")
print(result) # ['apple', 'banana', 'orange']

#문자열 결합
fruits = ["apple", "banana", "orange"]
result = ", ".join(fruits)
print(result)

#문자열 개수 세기
text = "banana"

print(text.count("a")) # 3
print(text.count("na")) # 2
```
