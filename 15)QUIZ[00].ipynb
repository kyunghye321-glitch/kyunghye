```python
# 실습2) 여러 공백을 하나의 공백으로 치환

import re

text = 'Python Data Analysis'
result = re.sub(r"\s+", '', text) # \s+ 1개 이상의 공백 문자
print(result)

import re

emails = [
    "python@네이버.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")

# 패턴이 안 맞아서 오류가 발생
```

```python
# 수량자 {n}

import re

text = "1,23,456,7890,77a8888"
print(re.findall(r"\d{3}", text))

# 숫자문자가 3번연속 나오는 구간만 반환 리스트에 추가
# ['456', '789', '888']

import re

text = '1 12 123 1234 12345'

print(re.findall(r"\d{2,4}", text))
```

```python
# OR( | ) 연산: 여러 패턴 중 하나와 일치하면 된다.

import re

text = "cat, dog, bird, cats, dog"

print(re.findall(r"cat|dog", text))
print(re.findall(r"(cat)|(dog)", text))
print(re.findall(r"[(cat)|(dog)]", text))
# ['cat', 'dog', 'cat', 'dog']
```

```python
import re

phones = [
    "02-1234-5678",
    "010-1234-5678",
    "010-123-5678",
    "010-1234-11111"
]

# "숫자3시작-숫자4-숫자4끝나는" 찾기
for p in phones:
    result = re.search(r"^\d{3}-\d{4}-\d{4}$", p)
    if result != None:
        print(result)
        print(result.group())
```

```python
# 실습4) URL 검증
# ^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$

import re

urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"
# ^https? - 'http' or 'https' 로 시작
# '://' = 'http://' or 'http://' 시작
# ([\w-]+ - []'\w' 이거나 '-']+ ex) http://abc, http://ab-c, http://://-a-
# ([\w-]+\.)+ - http://abc.abc.abc. www.naver.www.www
# (/[^\s]*)? - http://www.naver.www.www.com/

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

```python
# 실습5) 주민등록변호 형식 검사(외국사람 제외)

import re

nums = [
    "901010-1234567",
    "000101-4234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^\d{6}-[1-4]\d{6}$"

for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
