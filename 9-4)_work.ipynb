```python
data= []

file=open('9-4)_work.csv', 'r', encoding='utf-8')
lines = file. readlines()
```

```python
for line in lines[1:]:
    name,kor,eng=line.strip().split(",")

    data.append({
          "이름": name,
          "국어": kor,
          "영어": eng,
          })
    
file.close()

avg_list = list(map
    (lambda X: (x['이름'], (['국어'] + ['영어'] / 2 )),
data
))

sorted_avg = sorted(
    avg_list,
    key=lambda x: x[1],
    reverse=True
)

top_students = list(filter(
    lambda x: x[1] >=90,
    sorted_avg
))

print(sorted_avg)
print(top_students)
```
