```python
#딕셔너리(dict)
#테이블, 데이터프레임같은 2차원 데이터의 한행을 표현할 때 많이 사용

​# key : value 구조
# 순서 O (Python 3.7 이상)
# key는 중복 불가, value는 중복 가능
```

```python
#dict 자료형 생성, 다루기

# 빈 딕셔너리
d1 = {}
d2 = dict()

print(d1)
print(d2)

s = set()
```

```python
# {} 사용
student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}

print(student)
print(type(student))
```

```python
# {} 사용
student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}

print(student)
print(type(student))

# dict() 사용

person = dic(name="김철수", age=30, city="서울")
print(person)

# [()]를 dict으로 변환
data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]
fruit = dict(data)
print(fruit)

# 딕셔너리의 키(Key)는 반드시 해시 가능(Hashable)해야 한다.
# 그래서 불변(Immutable) 객체는 키로 사용할 수 있지만, 가변(Mutable) 객체는 키로 사용할 수 없다
```

```python
d = {"name":"홍길동"}
print(d)

d = {100: "학생"}
print(d)

d = {
    (1, 2): "좌표"
}
print(d)

d = {
    [1,2]: "리스트"
}   #error
print(d)
```

```python
# dict객체 연산, 메서드, 함수: <연산>

student = {"name": "홍길동"}
print(student)

# 요소조회
print(student["name"])

# 추가
student["age"] = 20
print(student)

# 수정
student["age"] = 30
print(student)

# 삭제
del student["age"]
print(student)

# in 연산
print("name" in student)
print("age" in student)
```

```python
# 메서드

student = {"name": "홍길동"}

print(student.get("name"))
print(student.get("age")) # 없으면 None 반환

student = {"name": "홍길동", "age": 20}
print(student.keys())
print(student.values())
print(student.items())

student.update({"age": 30, "score": 90})
print(student)

age = student.pop("age")
print(age)
print(student)

student.clear()
print(student)
```

```python
# 함수

student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}
print(len(student))

s = {
    "c": 3,
    "a": 1,
    "b": 2
}

print(sorted(s)) # 반환값은 정렬된 키의 리스트
print(s) # 원본 변화 없음

print(list(s)) # 반환값은 키의 리스트
```

```python
# 컬렉션객체

# 컬렉션객체는 for문을 사용하면 자신이 가진 아이템을 순환할 수 있다
# 문은 값을 반환하지 않는다
# 콜론문자(:)와 들여쓰기를 구문에 사용하여 함께 실행 되는 문들의 묶음인 코드 블록을 표현
# 문은 표현(식)을 포함할 수 있다. for문은 in연산 표현식이 사용
```

```python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)

colors = ("red", "green", "blue")

for color in colors:
    print(color)

nums = {10, 20, 30}

for num in nums:
    print(num)

student = {
    "name": "홍길동",
    "age": 20
}
# 키
for key in student:
    print(key)
# 값
for value in student.values():
    print(value)

# 키 + 값: 언패킹
for key, value in student.items():
    print(key, value)
```
