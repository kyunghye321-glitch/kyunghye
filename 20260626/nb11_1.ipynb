```python
# 조건표현식은 표현식, 조건문은 문장

# 조건표현식(삼항 연산자): 조건에 따라 하나의 값을 반환
age = 20
result = "성인" if age >= 19 else "미성년자"
# bool? 참값 : 거짓값
# a + b
#a++
print(result)

print(f"당신은 {'성인' if age >= 19 else '미성년자'} 입니다") # ""안에 문자열은 ''를 사용하자
```

```python
# 조건문(if 문)

# 문은 값을 반환하지 않는다.
# 콜론문자(:)와 들여쓰기를 구문을 사용하여 함께 실행 되는 문들의 묶음인 코드 블록을 표현

# if age >= 19       |       헤드라인(문법 키워드)
# :         |       코드 블록의 시작을 알리는 문법
# 들여쓰기      |       코드 블록에 속한 문들을 표시하는 문법
```
