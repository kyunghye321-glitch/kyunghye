```python
# 2. 알고리즘

# 1) 선형 탐색 - 반복 순환
# 실습 1. 선형 탐색 (10분)
# 주제: 원하는 숫자 찾기

# 리스트를 처음부터 끝까지 순차적으로 탐색한다.
# 데이터 개수가 증가할수록 탐색 횟수가 증가함을 확인한다.
# 데이터가 10배 증가하면 비교 횟수도 최대 10배 증가한다.

# 학습 개념
# - 순차 탐색
# - 시간복잡도 : O(n)

numbers = [15, 8, 23, 7, 31, 18]
target = 23

count = 0

for i in range(len(numbers)):
    count += 1
    if numbers[i] == target:
        print('찾은 위치 :', i)
        break # 찾으면 남은 원소는 비교하지 않고 break

print('비교 횟수: ', count)

# 최대값 찾기 - 반복 순환

numbers = [15, 8, 23, 7, 31, 18]

max_num = numbers[0]
count = 0

for n in numbers:
    count += 1
    if n > max_num:
        max_num = n

print('최댓값 :', max_num)
print('비교 횟수 :', count)
```

```python
# 덱 생성

from collection import deque

dq = deque()
dq.append(10)
dq.append(20)
dq.append(30) # 0번째 위치에 삽입

print(dq)

x = dq.pop() # 스텍처럼

print(x)
print(dq)

print(dq.popleft()) #큐처럼
print(dq)
```

```python
# 스텍이 필요: 효율성 Deque X -> list 사용
# 큐 필요 -> 효율성 0 -> Deque 사용
# 덱 필요 -> Deque 사용
# 자료구조의 회전 -> 시계 방향이냐, 반시계 방향이냐: 양방향 삽입/삭제 가능 -> 덱
```
