```python
def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True)

print(r)

r2 = sorted(s, key=score, reverse=False)
print(r2)

r2 = sorted(s, key=lambda e:e[1], reverse=False)
print(r2)
```

```python
#sorted: 정렬 기준 지정
names = ['banana', 'apple', 'kiwi']

# 문자열의 길이를 기준으로 오름차순 정렬
result = sorted(names, key=lambda x: len(x))
print(result)
```

```python
# stored()에 람다를 사용하면 데이터 과학에서 자주 사용되는 [dict, dict, ...] 자료구조를 원하는 키값으로 정렬이 가능
students = [
    {'이름': '홍길동', '나이': 20, '점수': 90},
    {'이름': '이순신', '나이': 18, '점수': 85},
    {'이름': '김유신', '나이': 22, '점수': 95},
]

# '나이'값을 기준으로 오름차순 정렬

result = sorted(student, key=lambda x: x['나이'])
print(result)

# x: 리스트의 각 딕셔너리
# x: ['나이']: 각 딕셔너리의 '나이' 값을 꺼냄
# sorted()는 이 값을 기준으로 정렬한다.
```

```python
# 여러 키를 기준으로 정렬 가능
students= [
        {'이름': '홍길동', '반': 20, '점수': 90},
        {'이름': '이순신', '반': 18, '점수': 85},
        {'이름': '김유신', '반': 22, '점수': 95},
]

# 1순위: 반, 2순위: 점수

result = sorted(students, key=lambda x: (x['반'], x['점수']))
print(result)
```

```python
#min(): 기준에 따라 최소값 찾기
names = ['banana', 'apple', 'kiwi']

# 문자열의 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))
print(result)
```
