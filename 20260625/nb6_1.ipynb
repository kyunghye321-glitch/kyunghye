```python
#복소수(complex)
#허수: 제곱이 음수가 되는 수

c = 3 + 4j

print(c.real) # 실수부 3.0
print(c.imag) # 허수부 4.0
```

```python
# 실수 (float)
# - 부동소수점 방식 사용 
# - 매우 큰 수와 작은 수 표현 가능 
# - 오차가 발생할 수 있음
# - 일반적인 수치 계산, 데이터 과학에서 사용

print(0.1 + 0.2) # 0.30000000000000004
```

```python

#Decimal 
# 고정밀 실수 계산
# Decimal이 10진수를 정확하게 저장하기 때문에 오차가 없다
# 금융 시스템, 회계 프로그램, 세금 계산, 정산 시스템 등에 사용
# 정확한 실수 연산이 필요한 핀테크 소프트웨어 개발시 사용(데이터 과학분야에서는 사용 빈도 낮음)

# decimal 모듈안에 Decimal함수를 메모리에 로드
# 파이썬 언어가 제공하는 객체중 내장객체가 아닌것들은 import 후 사용
from decimal import Decimal

a = Decimal("0.1") # 인자값을 float으로 전달하면 전달하는 순간 오차가 발생하니 문자열로 입력한다.
b = Decimal("0.2")

print(a + b)
print(0.1 + 0.2 == 0.3) # False
print(Decimal("0.1") + Decimal("0.2") == Decimal("0.3")) # True
```

```python
# Fraction
# 분수를 정확하게 표현
# 분수 연산이 필요한 수학에서 사용(데이터 과학분야에서는 사용 빈도 낮음)

from fractions import Fraction
f = Fraction(1, 3)

print(f)

# 분수계산의 예
a = Fraction(1, 3)
b = Fraction(1, 6)
c = a + b
print(c)

d = a*b
print(d)

# 분수를 실수로 변환
print(float(c))
print(float(d))
```

```python
# 숫자타입을 bool 타입으로 변환
# 0은 False, 0이 아닌 모든 수는 True
# bool()함수는 객체를 bool객체로 타입을 변경하는 함수

a = -1
b = 0
c = 1
d = 2
e = 0.1

print(bool(a)) # True
print(bool(b)) # False
print(bool(c)) # True
print(bool(d)) # True
print(bool(e)) # True
```

```python
# 숫자타입과 자주 사용되는 내장 함수
# 함수는 메서드와는 다르다. 메서드는 객체안에 포함되어 있는 그 객체만을 위한 기능이다.
# 함수는 어느 객체에 종속된 기능이 아닌 독립적인 코드(기능을 정의한)객체 이다.
# 숫자타입과 자주 사용되는 내장 함수는 숫자 타입을 인자값으로 사용하는 함수들이다.
# 숫자객체의 메서드들은 많지도 않고 사용 빈도도 낮다.

# ex)
# 메서드
# 1.some_fun() : 1이라는 숫자객체에만 어떠한 행위를 하는 기능(메서드)
# 함수
# some_fun(1) : 입력된 1이라는 숫자객체를 가지고 어떠한 행위를 하는 기능(함수)  

# 메소드 vs 함수
n = 10
print(n.bit_length()) # 10의 이진수 값(1010)의 길이

# bin()함수는 인자로 입력된 10진수의 2진수 문자열을 반환한다.
# 이진수 문자열 앞에 binary를 뜻하는 "0b"접두사를 앞에 붙여서 반환한다.
print(bin(n)) #0b1010 "0b" + "1010"

# bit_length()메서드는 n이라는 숫자 객체를 만들어야 존재하는 것이고,
# bin() 함수는 n객체와 상관없는 독립된 함수 객체이다.

print(bin(9)) # n과 상관없이 독립적으로 사용가능

# 숫자를 인자값으로 자주 사용되는 내장 함수

print(pow(2, 3)) # 거듭제곱 int로 반환
print(divmod(10, 3)) # 몫과 나머지를 tuple로 반환
print(round(3.14159, 2)) # 반올림된 실수값을 반환
print(abs(-10)) # 절대값을 반환(반환 타입은 인자값과 같은 타입)
```
