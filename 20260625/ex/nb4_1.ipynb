```python
# 함수 & 메서드
# 함수: 전역에서 정의된 독립적인 객체. 대부분 다른 객체를 매개값(인자)으로 받아 어떤 기능을 함.
# 메서드: 특정객체에 소속된 함수. 멤버함수 or 메서드라 부르며 해당객체에 대해서만 어떤 기능을 하는 함수

def fx_function():
    print('hello_python!')

fx_function()
fx_function()
fx_function()
fx_function()
# >> DRY(Don't Repeat Yourself)
```

```python
# dir 함수: dir함수는 인자로 전달된 객체에서 이용 가능한 모든 속성을 반환

s = "python"
print(dir(s))
print(s.upper())
print(s)
#대부분의 객체 메서드는 자기자신은 수정: 문자열은 불변이기 때문
```

```python
#대소문자 변환
text = "Python Programming"
print(text.upper())      # 모두 대문자
print(text.lower())      # 모두 소문자
print(text.capitalize()) # 첫 글자만 대문자
print(text.title())      # 각 단어 첫 글자 대문자

#공백제거
text = "   Python   "
print(text.strip())   # 양쪽 공백 제거
print(text.lstrip())  # 왼쪽 공백 제거
print(text.rstrip())  # 오른쪽 공백 제거
```

```python
#문자열 판별: f-string 주로 사용
print("123".isdigit())      # 문자열이 숫자로만 구성되어 있는가?
print("Python".isalpha())   # 문자열이 영문자(또는 한글 등 문자)로만 구성되어 있는가?
print("Python123".isalnum()) # 문자열이 문자와 숫자로만 구성되어 있는가?

#문자열 채우기

num = "100"

print(num.zfill(5))    # 전체 길이가 5가 되도록 왼쪽을 0으로 채움
print(num.center(10))  # 전체 길이가 10이 되도록 가운데 정렬
print(num.ljust(10))   # 전체 길이가 10이 되도록 왼쪽 정렬
print(num.rjust(10))   # 전체 길이가 10이 되도록 오른쪽 정렬

#전화번호 연결
phone1 = "010"
phone2 = "1234"
phone3 = "5678"
phone = "-".join([phone1, phone2, phone3])
print(phone)

#이메일에서 아이디 추출
email = "python@test.com"
user_id = email.split("@")[0] #  # @를 기준으로 분리한 후 첫 번째 요소(인덱스 0) 추출
print(user_id)
```

```python
#포맷 지정 출력
#f-string(가장 많이 사용)

name = "Park"
age = 30

# f-string 기본 사용
print(f"이름: {name}, 나이: {age}")  # 변수 값을 문자열 안에 직접 삽입

# 소수점 자리수 지정
pi = 3.141592
print(f"{pi:.2f}")  # 소수점 둘째 자리까지 출력 → 3.14

# 자리수(폭) 지정
num = 7
print(f"{num:5}")   # 전체 5칸 확보, 오른쪽 정렬 → '    7'
print(f"{num:05}")  # 빈칸을 0으로 채움 → 00007
```

```python
##format() 함수
# 기본 사용
print("이름: {}, 나이: {}".format("Park", 30))  # {} 순서대로 값 들어감

# 인덱스 사용
print("이름: {0}, 나이: {1}".format("Park", 30))  # 번호로 위치 지정

# 이름 지정
print("이름: {name}, 나이: {age}".format(name="Park", age=30))  
# 변수 이름으로 직접 지정 가능
```

```python
## % 연산자 
name = "Park"
age = 30

print("이름: %s, 나이: %d" % (name, age))  
# %s = 문자열, %d = 정수

pi = 3.141592
print("%.2f" % pi)  # 소수점 둘째 자리까지 출력 → 3.14
```

```python
# 문자열 정렬
# 왼쪽 정렬 (<), 오른쪽 정렬 (>), 가운데 정렬 (^)
print(f"{'Python':<10}")  # 왼쪽 정렬
print(f"{'Python':>10}")  # 오른쪽 정렬
print(f"{'Python':^10}")  # 가운데 정렬
```
