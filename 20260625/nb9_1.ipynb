```python
nums = {1, 2, 3, 3, 3}
print(nums) # {1, 2, 3}

a = {1, 2, 3}
b = {3, 4, 5}

# 교집합
print(a & b)
print(a.intersection(b))

#합집합
print(a | b)            # {1, 2, 3, 4, 5}
print(a.union(b))

#차집합
print(a - b)            # {1, 2}
print(a.difference(b))
```

```python
s = {1, 2, 3}
s.add(4)        # 하나 추가
print(s)

s.update([5, 6])    #여러 개 추가
print(s)

s.remove(2)         # 값 2 삭제 (없으면 오류)
print(s)

s.discard(10)       # 값 없으면 그냥 무시
```
