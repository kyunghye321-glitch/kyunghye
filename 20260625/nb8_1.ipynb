```python
# 리스트: 다른 객체들을 모아 그룹으로 처리할 수 있는 정렬된 컬렉션
# 리스트에 포함된 객체 아이템은 자신만의 위치를 가진다. 인덱싱이나 슬라이싱같은 작업을 수행할 수 있다.
# 어떠한 객체도 포함할 수 있다. 리스트안에 리스트객체를 가지는 중첩도 가능.
# 가변 타입이며, 크기를 늘리거나 줄일 수 있다. 
# 문자열(불변)이 지원하지 않는 삭제, 인덱스 할당연산 같이 리스트를 직접변경하는 연산들도 지원한다.
```

```python
# 리스트의 연산

# 1) 연결(+) 연산
a = [1, 2, 3]
b = [4, 5]
result = a + b
print(result) # [1, 2, 3, 4, 5]

a = a + b
print(a)

c = "py"
c = c + "thon"
print(c)

# 2) 반복(*) 연산
nums = [10, 20]
result = nums * 3
print(result) # [10, 20, 10, 20, 10, 20]

# 3) in, not in 연산
fruits = ["apple", "banana", "orange"]
print("banana" in fruits)

print("grape" in fruits)
print("grape" not in fruits)

# 4) 인덱싱
nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1])

# 인덱스 할당 연산
nums[2] = 99 # 리스트는 가변 객체
print(nums)

# 5) 슬라이싱 : 슬라이싱 연산은 리스트 자신이 변경되는 게 아니고 슬라이싱 된 리스트를 반환

nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50] 슬라이싱 연산 후에도 리스트 자신은 변경되지 않는다. 

# 2칸씩 건너뛰기
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::2])

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[1::2])

# 역순 출력
nums = [10, 20, 30, 40, 50]
print(nums[::-1]) # [50, 40, 30, 20, 10]

# 뒤에서 3개 추출
nums = [10, 20, 30, 40, 50, 60]
print(nums[-3:])

# 뒤에서 4개 중 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-4::2])

# 뒤에서 앞으로 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::-2])

# [5]위치부터 뒤로(-1) [1]위치 전까지
nums = [10, 20, 30, 40, 50, 60, 70]
print(nums[5:1:-1]) # 5위치부터 1전까지 -1칸식  [60, 50, 40, 30]

# 마지막 5개를 역순으로
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-1:-6:-1])

# 슬라이싱 할당(슬라이싱 구간에 대입문)
nums = [10, 20, 30, 40, 50]
nums[1:4] = [7, 8] # 1~3구간 [20, 30, 40]에 [7, 8]을 대입 
print(nums) # [10, 7, 8, 50]

# 슬라이싱 결과를 다른 변수에 대입
a = [10, 20, 30, 40, 50, 60]
b = a[:]
c = a[::-1]
d = a[1::2]
e = a[:2]

print(b)
print(c)
print(d)
print(e)
```
