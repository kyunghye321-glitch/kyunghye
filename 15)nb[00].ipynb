```python
import re

text = "apple banana orange"
result = re.findall(r"a", text)
print(result) # ['a', 'a', 'a', 'a', 'a']
```

```python
import re

text = "A10 B20 C30"
numbers = re.findall(r"\d+", text) # \d 숫자한문자, \d+ 연속된 숫자 한문자이상
print(numbers)
```

```python
# 숫자 문자 제거
import re

text = "A10 B20 C30"
result = re.sub(r"\d", "", text)
print(result)

import re

text = 'A10 B20 C30'
# result = re.sub(r"\d, '', text)
result = re.sub(r"[r^\d]", '0', text)
print(result)
```
