```python
# 8 

num = int(input('정수: '))
prime_N = True
if num <= 1:
    prime_N = False
else:
    for i in range(2, num):
        if num % i == 0:
            prime_N = False
            break

if prime_N:
    print('소수')
else:
    print('소수 X')
```

```python
# 9

count = 0

for num in range(3, 101):
    prime_N = True

    for i in range(2, num):
        if num % i == 0:
            prime_N = False
            break

    if prime_N:
        count += 1


print(count)
```
