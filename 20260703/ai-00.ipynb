```python
# 8 

num = int(input('정수: '))
prime_N = True
if num <= 1:
    prime_N = False
else:
    for i in range(2, num):
        if num % i == 0:
            prime_N = False
            break

if prime_N:
    print('소수')
else:
    print('소수 X')
```

```python
# 9

count = 0

for num in range(3, 101):
    prime_N = True

    for i in range(2, num):
        if num % i == 0:
            prime_N = False
            break

    if prime_N:
        count += 1


print(count)
```

```python
# 11

def print_x():
    n = 10

for i in range(n):
    if i < 5:
        left = i
        mid = 8 - 2*i
    else:
        left = n - i - 1
        mid = 2*(i-5)
    print(' ' * left + '*' + ' ' * mid + '*')
```
