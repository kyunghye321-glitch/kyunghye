```python
"\\d" -> r"\d"

print("Hello\nworld\\")
# 출력
# Hello
# world\

print(r"Hello\nworld\\")
# 출력 
# Hello\nworld\\

import re

text = "apple banana cat"
result = re.findall(r"[abc]", text)
# "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가

print(result)
# ['a', 'b', 'a', 'a', 'a', 'c', 'a']
```

```python
# 영문대문자 찾기

import re

text = "Python JAVA AI"
print(re.findall(r"[A-Z]", text))
```

```python
# 대괄호[] 안의 "-"는 범위를 나타낸다.

import re

text = "A12B34"
print(re.findall(r"[0-9]", text))
# ['1', '2', '3', '4']
# 문자0부터 문자9까지 하나라도 일치하면 반환되는 리스트에 추가 
# 정규표현식 [0-9]는 [0123456789] 동일한 패턴
```

```python
# 소대문자 포함, 특수문자 포함, 숫자문자 포함

import re
password = input('비밀번호:')
result = []

if not re.search(r'[A-Z]', password):
    result.append('대문자가 없습니다')
if not re.search(r'[0-9]', password):
    result.append('숫자가 없습니다')
if not re.search(r'[a-z]', password):
    result.append('소문자가 없습니다')
if not re.search(r'[^0-9a-zA-Z]', password):
    result.append('특수문자가 없습니다')
if len(result) == 0:
    result.append('사용 가능한 비밀번호입니다.')

result
```

```python
import re

text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
import re

text = "cat ct cot cut coot"
result = re.findall(r"c.t", text) # "c_t" 이런 패턴의 문자(c로 시작, t로 끝나는 문자열길이 3) 찾기
print(result)
```

```python
import re

# findall함수에서는 ^패턴을 사용해도 의미있는 반환값을 얻을 수 없다(^패턴은 전체 문자열을 하나로 다루기 때문)
text = "Python Python python"
print(re.findall(r"^Py", text)) # ["Py"]

text = "python Python python"
print(re.findall(r"^Py", text)) # [] 
```

```python
import re

text = "Python Python python"
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.search(r"^py", text)) # None
```

```python
import re

text = "Python"

print(re.search(r"hon$", text)) # <re.Match object; span=(3, 6), match='hon'>
print(re.search(r"abc$", text)) # None
```

```python
import re

text = "Pyyyyython"

#시작이 Py 찾기
print(re.search(r"^Py", text)) # <re.Match object; span=(0,2), match='Py'>
# 시작이 Pyyyy... 찾기
print(re.search(r"Py+", text)) # <re.Match object; span=(0,6), match='Pyyyyy'>

import re

texts = ['3', '123,', '', 'abc', '12a']

for t in texts:
    print(t, '->', bool(re.fullmatch(r'\d*', t))) # \d* : \d+ : 숫자가 1개 이상 있어야 한다.
```

```python
import re

words = [
    'ae'
    'abc'
    'abbcd'
    'abbbcde'
    'acdc'
]

# ab?c: b문자가 0개 or 1개 패턴

for word in words:
    print(re.match(r'ab?c', word))

# None
# <re.Match object; span=(0,3), match='abc'>
# None
# None
# <re.Match object; span=(0,3), match='ac'>
```

```python
import re

text = "Py@thon 123_xyz"

print(re.findall(r"\d", text))
print(re.findall(r"\D", text))
print(re.findall(r"\w", text))
print(re.findall(r"\W", text)) # 특수문자, 공백
print(re.findall(r"\s", text)) # 공백
```
