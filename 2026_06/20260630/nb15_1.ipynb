```python
import csv

# open()은 내장함수
with open('students.csv', 'r', encoding='utf-8') as f:
    # reader(): csv모듈안에 있는 함수
    reader = csv.reader(f)
    print(type(reader))
    for row in reader:
        print(row)
        print(type(row)) # <class 'list'>
```

```python
with open('student.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f) # 헤드라인 키, 한줄의 내용은 값

    next(reader)

    for row in reader:
        print(row)
```
