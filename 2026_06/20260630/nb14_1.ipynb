```python
# 루프 else

# break가 없는 경우
for i in range(3):
    print(i)
else:
    print('정상 종료')

# break가 있는 경우
for i in range(3):
    if i == 1:
        break
    print(1)
else:
    print('정상 종료')
# break가 실행되었으므로 else는 실행되지 않는다.
```

```python
# while문에서도 동일하게 사용 가능

i = 1

while i <= 5:
    if i == 3:
        break
    print(i)
    i += 1

print('반복 종료')
```

```python
# try-except문
# 예외(Exception)가 발생했을 때 프로그램이 종료되지 않고 원하는 처리 수행
# 예외는 코드 에러는 아니지만 프로그램 실행 중 사용자 입력값, 실행환경 등에 의해 발생하는 오류

num = input('숫자를 입력하세요')
print(num*10)
```

```python
try:
    num = int(input('숫자 입력 : ')) **2
    print(num)
except:
    print('숫자가 아닙니다.')

print('다음 실행문...')
```

```python
try:
    num = int(input('숫자 입력 : '))
    result = 10 / num
    print(result)

except ValueError:
    print('숫자를 입력하세요')

except ZeroDivisionError:
    print('0으로 나눌 수 없습니다.')
```

```python
try:
    num = int(input("숫자 입력: "))
    result = 10 / num
    print(result)


except:
    print('자원 해지')

finally:
    print('프로그램 종료')

print('종료')
```

```python
# try:
#     # 예외가 발생할 수 있는 코드

# except 예외종류:
#     # 예외 처리

# else:
#     # 예외가 없을 때 실행 (선택)

# finally:
#     # 항상 실행 (선택)
```

```python
try:
    num = int(input('숫자 입력: '))
    result = 100 / num

except ValueError:
    print('숫자를 입력해야합니다')

except ZeroDivisionError:
    print('0으로 나눌 수 없습니다')

else:
    print('계산 결과:', result)

finally:
    print('프로그램 종료')
```
