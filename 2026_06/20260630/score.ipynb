```python
import csv

with open('score.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    data = list(reader)

print(data)
```

```python
for student in data:
    print(student)
```

```python
for student in data:
    if student['국어'] in ['', '결측', 'NaN']:
        student['국어'] = '0'

    if student['영어'] in ['', '결측', 'NaN']:
        student['영어'] = '0'

    if student['나이'] in ['', '결측', 'NaN']:
        student['나이'] = '0'


for student in data:
    print(student)
 
```

```python
for student in data:
    student['국어'] = int(student['국어'])
    student['영어'] = int(student['영어'])
    student['나이'] = int(student['나이'])
```

```python
for student in data:

    if student['나이'] > 100:
        student['나이'] = 0

    if  student['국어'] < 0 or student['국어'] > 100:
        student['국어'] = 0

    if student['영어'] < 0 or student['영어'] > 100:
        student['영어'] = 0
```

```python
for student in data:
    student['총점'] = student['국어'] + student['영어']

for student in data:
    student['평균'] = student['총점'] /2
```

```python
for student in data:

    if student['평균'] >= 60:
        student['결과'] = '합격'
    else:
        student['결과'] = '불합격'
```

```python
for current_student in data:

    rank = 1

    for other_student in data:
        if other_student['총점'] > current_student['총점']:
            rank += 1

    current_student['순위'] = rank
```

```python
total = 0

for student in data:
    total += student['국어']

print('국어 평균:', total / len(data))
```

```python
total = 0

for student in data:
    total += student['영어']

print('영어 평균:', total / len(data))
```

```python
max_value = data[0]['평균']

for student in data:
    if student['평균'] > max_value:
        max_value = student['평균']

print('최고 평균:', max_value)
```

```python
min_value = data[0]['평균']

for student in data:
    if student['평균'] < min_value:
        min_value = student['평균']

print('최저 평균:', min_value)
```

```python
pass_count = 0

for student in data:
    if student['결과'] == '합격':
        pass_count += 1

print('합격률:', pass_count / len(data) * 100, '%')
```

```python
kor_total = 0
eng_total = 0
pass_count = 0
max_avg = data[0]['평균']
min_avg = data[0]['평균']

for student in data:

    kor_total += student['평균']
    eng_total += student['평균']

    if student['결과'] == '합격':
        pass_count += 1

    if student['평균'] > max_avg:
        max_avg = student['평균']

    if student['평균'] < min_avg:
        min_avg = student['평균']

print('국어 평균:',kor_total / len(data))
print('영어 평균:',eng_total / len(data))
print('최고 평균:', max_avg)
print('최저 평균:', min_avg)
print('합격률:', pass_count / len(data) * 100, '%')
```
