```python
# 리스트: 다른 객체들을 모아 그룹으로 처리할 수 있는 정렬된 컬렉션
# 리스트에 포함된 객체 아이템은 자신만의 위치를 가진다. 인덱싱이나 슬라이싱같은 작업을 수행할 수 있다.
# 어떠한 객체도 포함할 수 있다. 리스트안에 리스트객체를 가지는 중첩도 가능.
# 가변 타입이며, 크기를 늘리거나 줄일 수 있다. 
# 문자열(불변)이 지원하지 않는 삭제, 인덱스 할당연산 같이 리스트를 직접변경하는 연산들도 지원한다.
```

```python
# 리스트의 연산

# 1) 연결(+) 연산
a = [1, 2, 3]
b = [4, 5]
result = a + b
print(result) # [1, 2, 3, 4, 5]

a = a + b
print(a)

c = "py"
c = c + "thon"
print(c)

# 2) 반복(*) 연산
nums = [10, 20]
result = nums * 3
print(result) # [10, 20, 10, 20, 10, 20]

# [30, 60] NumPy라이브러리로 행렬객체를 만들어 사용

# 3) in, not in 연산
fruits = ["apple", "banana", "orange"]
print("banana" in fruits)

print("grape" in fruits)
print("grape" not in fruits)

# 4) 인덱싱
nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1])

# 인덱스 할당 연산
nums[2] = 99 # 리스트는 가변 객체
print(nums)

n2 = (3, 5)
print(n2[0])


# 5) 슬라이싱 : 슬라이싱 연산은 리스트 자신이 변경되는 게 아니고 슬라이싱 된 리스트를 반환

nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50] 슬라이싱 연산 후에도 리스트 자신은 변경되지 않는다. 

# 2칸씩 건너뛰기
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::2])

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[1::2])

# 역순 출력
nums = [10, 20, 30, 40, 50]
print(nums[::-1]) # [50, 40, 30, 20, 10]

# 뒤에서 3개 추출
nums = [10, 20, 30, 40, 50, 60]
print(nums[-3:])

# 뒤에서 4개 중 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-4::2])

# 뒤에서 앞으로 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::-2])

# [5]위치부터 뒤로(-1) [1]위치 전까지
nums = [10, 20, 30, 40, 50, 60, 70]
print(nums[5:1:-1]) # 5위치부터 1전까지 -1칸식  [60, 50, 40, 30]

# 마지막 5개를 역순으로
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-1:-6:-1])

# 슬라이싱 할당(슬라이싱 구간에 대입문)
nums = [10, 20, 30, 40, 50]
nums[1:4] = [7, 8] # 1~3구간 [20, 30, 40]에 [7, 8]을 대입 
print(nums) # [10, 7, 8, 50]

# 슬라이싱 결과를 다른 변수에 대입
a = [10, 20, 30, 40, 50, 60]
b = a[:]
c = a[::-1]
d = a[1::2]
e = a[:2]

print(b)
print(c)
print(d)
print(e)
```

```python
a = [10, 20, 30]
b = a 

print(a)
print(b)
print(a == b)
print(a is b)

b[0] = 999
print(a) 
print(b) 
```

```python
a = [10, 20, 30]
b = a[:]

print(a)
print(b)
print(a == b) 
print(a is b)

b[0] = 999
print(a) # [10, 20, 30]
print(b) # [999, 20, 30]
```

```python
# * 자주사용하는 슬라이싱

# data[:]      | 전체
# data[:5]     | 앞 5개
# data[5:]     | 5번째 이후
# data[-5:]    | 마지막 5개
# data[::-1]   | 역순
```

```python
# 리스트 비교연산
a = [1, 2, 3]
b = [1, 2, 3]
c = [1, 2, 4]

print(a == b) # True
print(a == c) # False
```

```python
# 리스트를 인자값으로 자주 사용되는 내장함수

nums = [10, 20, 30, 40]

print(len(nums))
nums = [10, 20, 30, 40]

print(max(nums))
print(min(nums))
print(sum(nums))
```

```python
# 리스트 메서드: 리스트 객체는 자신만의 메서드를 가지며, 대부분의 리스트 메서드는 자기 자신을 수정하는데 사용

# append() : 맨 뒤에 1개 추가
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)

# insert() : 원하는 위치에 추가
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)

# extend() : 여러 개 추가
nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)

# remove() : 값으로 삭제
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)

# 수정은 인덱스 할당, 슬라이싱 할당으로 수정

# pop() : 인덱스로 삭제 후 반환
nums = [10, 20, 30]
result = nums.pop() # 마지막 인덱스 삭제 후 반환
print(result)
print(nums)

nums = [10, 20, 30]
result = nums.pop(1) # 인덱스1 삭제 후 반환
print(result)
print(nums)

# clear() : 전체 삭제
nums = [10, 20, 30]
nums.clear()
print(nums)
```

```python
# 삭제는 del문 사용 가능
# del 문
nums = [10, 20, 30, 40]
del nums[1]
print(nums)

# reverse() : 순서 뒤집기, 정렬과 상관없이 현재 순서를 거꾸로 만든다.
nums = [10, 20, 30, 40]
nums.reverse()
print(nums)

# count() : 특정 값이 몇개인지
nums = [10, 20, 10, 30, 10]
print(nums.count(10)) # 3

# index() : 특정 값의 인덱스
fruits = ["apple", "banana", "orange"]
print(fruits.index("banana")) # 1
```

```python
# 리스트 정렬

# 정렬 메서드
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)

# 정렬의 경우 내장함수를 사용하여도 된다.
# sorted 내장함수는 리스트 뿐만 아니라 모든 컬렉션 타입을 정렬할 수 있고, 원본은 변경없이 정렬된 객체를 반환

nums = [50, 10, 40, 20]
result = sorted(nums)
print(nums)
print(result)

nums = [50, 10, 40, 20]
result = sorted(nums, reverse=True)
print(result)

fruits = ["banana", "apple", "orange"]
result = sorted(fruits)
print(result)

# key매갲변수를 이용하여 문자열 길이로 정렬
fruits = ["watermelon", "kiwi", "banana", "apple"]
result = sorted(fruits, key=len)
print(result)

# 매개변수 key값으로는 함수가 온다.
# 뒤에서 람다함수를 배우면 # key값으로 람다식을 대입하여 다양한 요구 사항의 정렬도 가능
# 자기 자신을 정렬할때는 sort()메서드, 다양한 형태의 정렬된 리스트를 반환할때는 sorted() 내장함수 사용
```
