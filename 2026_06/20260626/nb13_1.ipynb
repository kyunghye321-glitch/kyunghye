```python
# 컬렉션 타입(객체)의 복사

a = [10, 20, 30]
b = a.copy()

print(a == b)
print(a is b)

a = {10, 20, 30}
b = a.copy()

print(b)
```

```python
student = {
    "name": "홍길동",
    "age": 20
}

student2 = student.copy()

print(student2)
```

```python
# 튜플은 불변객체라 할당하면 자동으로 복사

a = (10, 20, 30)
b = a

print(a)
print(b)
```

```python
​# 중첩된 컬렉션의 복사시 유의 사항

​# 객체의 copy()메서드는 현재 객체만 복사하는 얕은 복사(Shallow Copy)가된다.
# 바깥 리스트는 새로 생성됨
# 내부 리스트는 copy()를 사용하지 않고 할당(공동 참조)을 한다. 
# 내부 리스트의 참조(주소)만 복사
```

```python
# Shallow Copy(얕은 복사)는 "1단계만 복사"하고 내부 객체의 주소(참조)는 그대로 복사
# 그래서 중첩된 리스트·딕셔너리 등은 공동 참조가 발생

a = [[1, 2], [3, 4]]
b = a.copy()

# 바깥 리스트는 다른 객체
print(a is b)      # False

# 내부 리스트는 같은 객체
print(a[0] is b[0]) # True
```

```python
a = [[1, 2], [3, 4]]
b = a.copy()

b[0][0] = 999 # 내부리스트는 공동참조

print(a) # [[999, 2], [3, 4]]
print(b) # [[999, 2], [3, 4]]

b[1] = 7 # 바깥 리스트는 다른 객체
print(a) # [[999, 2], [3, 4]]
print(b) # [[999, 2], 7]
```

```python
# 중첩 객체까지 모두 복사할 때에는 copy모듈의 deepcopy()함수를 사용하여 깊은 복사(Deep Copy)를 해야 한다.

import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

b[0][0] = 999

print(a)
print(b)
```
