```python
students = [
    {"이름": "홍길동", "점수": 90},
    {"이름": "김철수", "점수": 85},
    {"이름": "이영희", "점수": 95}
]
```

```python
# 스키마(구조)

# 이름                  점수
# 홍길동                 90
# 김철수                 85
# 이영희                 95

# [] : list를 이용하여 여러 행(row)을 저장 
# {} : dict을 이용하여 한 행의 열(column) 데이터를 저장
```

```python
# 데이터 분석 실습
# [데이터 수집]
# 1) 데이터 준비

students = [
    {"이름": "김철수", "성적": 95, "주민번호": "980101-1234567"},
    {"이름": "이영희", "성적": 88, "주민번호": "010305-4234567"},
    {"이름": "박민수", "성적": None, "주민번호": "950720-1234567"},  # 결측치
    {"이름": "최지은", "성적": 77, "주민번호": "030215-4234567"},
    {"이름": "정우성", "성적": 105, "주민번호": "990101-1234567"},   # 이상치
    {"이름": "한지민", "성적": 82, "주민번호": "020912-4234567"},
    {"이름": "강호동", "성적": -10, "주민번호": "700511-1234567"},   # 이상치
    {"이름": "신세경", "성적": 91, "주민번호": "971224-2234567"},
    {"이름": "유재석", "성적": 85, "주민번호": "720814-1234567"},
    {"이름": "송혜교", "성적": None, "주민번호": "811122-2234567"}, # 결측치
    {"이름": "차은우", "성적": 79, "주민번호": "970330-1234567"},
]

# print(students) 보다 students 출력하는 것이 Jupyter Notebook에서 가독성이 좋다
```

```python
# [데이터 전처리]
# 2) 결측치 확인

missing_count = 0 # 결측치가 몇 개인지?

for student in students:
    if student["성적"] is None:
        missing_count += 1

print("결측치 개수:", missing_count)
```

```python
# 3) 결측치 처리(평균값 대체 : 어떤 값으로 대처하는지는 요구사항에 따라 다름)

scores = []

for student in students:
    if student["성적"] is not None:
        scores.append(student["성적"])
```

```python
avg = sum(scores) / len(scores) #None을 제외한 성적의 평균

for student in students:
    if student["성적"] is None:
        student["성적"] = avg

print(students)

students
#쥬피트 노트북에서는 print(students)로 출력하는 것보다 변수이름으로 출력하는 것이 출력결과의 가독성이 좋음
```

```python
# 4) 이상치 탐지(정상 범위 : 0 ~ 100점)
for student in students:
    score = student["성적"]

    if score < 0 or score > 100:
        print(student)
```

```python
# 5) 이상치 처리(어떻게 처리할지는 요구사항에 따라 다름)
for student in students:

    if student["성적"] > 100:
        student["성적"] = 100

    elif student["성적"] < 0:
        student["성적"] = 0

students
```

```python
# 구분                    특성 추출(Feature Extraction)                       파생변수 생성(Feature Engineering)
# 의미                    기존 데이터에서 의미 있는 정보를 뽑아냄                 기존 데이터를 가공하여 새로운 변수 생성    
# 결과                    기존 정보의 일부를 추출                               새로운 정보나 변수 생성
# 예시                    주민번호 → 성별                                      키, 몸무게 → BMI
# 예시                    날짜 → 연도, 월                                     구매금액 → 구매등급
# 예시                    주소 → 시/도                                        나이 → 연령대
```

```python
# 성별 구하기(특성 추출)

for student in students:

    gender_code = student["주민번호"].split("-")[1][0]

    if gender_code in ["1", "3"]:
        student["성별"] = "남"
    else:
        student["성별"] = "여"

students
```

```python
# 나이 구하기(특성 추출)

from datetime import datetime

current_year = datetime.now().year

for student in students:
    rrn = student["주민번호"]
    yy = int(rrn[:2])
    code = rrn.split("-")[1][0]

    if code in ["1", "2"]:
        birth_year = 1900 + yy
    else:
        birth_year = 2000 + yy

    student["나이"] = current_year - birth_year

students
```

```python
# [EDA : 탐색적 데이터 분석]
# 기술통계

scores = []

for student in students:
    scores.append(student["성적"])

print("개수:", len(scores))
print("합계:", sum(scores))
print("평균:", sum(scores) / len(scores))
print("최대:", max(scores))
print("최소:", min(scores))
```

```python
# 성별 평균 성적

male_sum = 0
male_count = 0

female_sum = 0
female_count = 0

for student in students:

    if student["성별"] == "남":
        male_sum += student["성적"]
        male_count += 1

    else:
        female_sum += student["성적"]
        female_count += 1

print("남 평균:", male_sum / male_count)
print("여 평균:", female_sum / female_count)
```

```python
# 연령대별 평균 성적

age_group = {}

for student in students:

    decade = (student["나이"] // 10) * 10

    if decade not in age_group:
        age_group[decade] = []

    age_group[decade].append(student["성적"])

print(age_group)
```

```python
for decade, scores in age_group.items():
    avg = sum(scores) / len(scores)
    print(f"{decade}대 평균: {avg:.2f}")
```
